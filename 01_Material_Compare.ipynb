```python
from mp_api.client import MPRester
import pandas as pd

API_KEY = "TYPE_YOUR_API_KEY_HERE" # 這是你截圖中的 Key

with MPRester(API_KEY) as mpr:
    # 1. 這裡把 num_results 或 limit 都刪掉，只保留搜尋條件
    docs = mpr.summary.search(
        elements=["Li", "Fe", "O"],
        is_stable=True,
        fields=["formula_pretty", "band_gap", "density", "energy_above_hull", "formation_energy_per_atom"]
    )
    
    data_list = []
    
    # 2. 在這裡加上 [:10]，代表只取搜尋結果的前 10 筆
    for doc in docs[:10]:
        data_list.append({
            "化學式": doc.formula_pretty,
            "能隙 (eV)": doc.band_gap,
            "密度 (g/cm³)": doc.density,
            "穩定性 (E_hull)": doc.energy_above_hull,
            "形成能 (eV/atom)": doc.formation_energy_per_atom
        })

df = pd.DataFrame(data_list)
print(df.head()) # 先印出來看看有沒有成功

# 儲存檔案
df.to_csv("my_materials.csv", index=False)

most_stable = df.loc[df['穩定性 (E_hull)'].astype(float).idxmin()]
print(f"最推薦研究的穩定材料是: {most_stable['化學式']}")
```
